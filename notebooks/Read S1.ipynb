```python
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
def load_html_data_to_dataframe(folder_location):
    list_of_html = []
    for file in os.listdir(folder_location):
        with open(f"{folder_location}/{file}", "r") as document_reader:
            tmp_html = document_reader.read()
            list_of_html.append(tmp_html)
    dataframe = pd.DataFrame(list_of_html, columns=["HTML"])
    return dataframe
```

```python
def search_for_lock_period(html_text):
    keywords = ["lock-up", "lock up",]
    tmp_soup_obj = BeautifulSoup(html_text, "lxml")
    lockup_hits = tmp_soup_obj.find_all(lambda tag: tag.name == "tr" and any(keyword in tag.text.strip().lower() for keyword in keywords))
    if len(lockup_hits) >= 0:
        result = " ".join([hit.text for hit in lockup_hits])
    else:
        result = "No hits."
    return result
```

```python
data_location = "../data"
df = load_html_data_to_dataframe(data_location)
```

```python
df["Lockup"] = df["HTML"].apply(search_for_lock_period)
```
